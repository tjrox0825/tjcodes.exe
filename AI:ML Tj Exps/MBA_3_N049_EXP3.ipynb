```
data=pd.read_csv("sat_cgpa.csv")
data.head()
data.describe()
```

```
y=data["GPA"]
x1=data["SAT"]
plt.scatter(x1,y)
plt.xlabel("SAT",fontsize=15)
plt.ylabel("GPA",fontsize=15)
plt.title("Data Points",fontsize=30,color="green")
```

```
x=sm.add_constant(x1)
results=sm.OLS(y,x).fit()
results.summary()
```

```
plt.scatter(x1,y)
yhat=0.2750+(0.0017*x1)
plt.plot(x1,yhat,lw=4,c='orange')
plt.xlabel('SAT',fontsize=15)
plt.ylabel('GPA',fontsize=15)
plt.title('Regression Line',fontsize=30,color='green')
```
