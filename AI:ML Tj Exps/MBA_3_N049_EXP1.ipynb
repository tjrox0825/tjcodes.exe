```
marks=input("Enter the marks")
def calc(marks):
  if marks<"35":
    return "Fail"
  elif marks<"60" and marks>"34":
    return "3rd class"
  elif marks<"80" and marks>"59":
    return "2nd class"
  else:
    return "1st class"
ans=calc(marks)
print(ans)
```
