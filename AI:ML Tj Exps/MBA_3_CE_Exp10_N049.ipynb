```
#Experiment Number 10 
# K-means algorithm 
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```
#create dataset 
x,y=make_blobs(n_samples=150,n_features=2,centers=3,cluster_std=0.8,shuffle=True,random_state=0)
#y.shape 
plt.scatter(x[:,0],x[:,1],c='white',marker='o',edgecolor='black',s=50)
```

```
from sklearn.cluster import KMeans
```

```
km=KMeans(n_clusters=3, init='random',max_iter=1,tol=0.001,random_state=0)
k1=km.fit_predict(x)
```

```
plt.scatter(x[k1==0,0],x[k1==0,1],s=50,c='green',marker='o',label='cluster 1')
plt.scatter(x[k1==1,0],x[k1==1,1],s=50,c='yellow',marker='s',label='cluster 2')
plt.scatter(x[k1==2,0],x[k1==2,1],s=50,c='blue',marker='v',label='cluster 3')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],s=100,c='red',marker='*',label='centroids')
plt.legend(scatterpoints=1)
km.cluster_centers_
```

```
km=KMeans(n_clusters=3, init='random',max_iter=3,tol=0.001,random_state=0)
k1=km.fit_predict(x)
plt.scatter(x[k1==0,0],x[k1==0,1],s=50,c='green',marker='o',label='cluster 1')
plt.scatter(x[k1==1,0],x[k1==1,1],s=50,c='yellow',marker='s',label='cluster 2')
plt.scatter(x[k1==2,0],x[k1==2,1],s=50,c='blue',marker='v',label='cluster 3')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],s=100,c='red',marker='*',label='centroids')
plt.legend(scatterpoints=1)
km.cluster_centers_
```

```
km=KMeans(n_clusters=3, init='random',max_iter=10,tol=0.001,random_state=0)
k1=km.fit_predict(x)
plt.scatter(x[k1==0,0],x[k1==0,1],s=50,c='green',marker='o',label='cluster 1')
plt.scatter(x[k1==1,0],x[k1==1,1],s=50,c='yellow',marker='s',label='cluster 2')
plt.scatter(x[k1==2,0],x[k1==2,1],s=50,c='blue',marker='v',label='cluster 3')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],s=100,c='red',marker='*',label='centroids')
plt.legend(scatterpoints=1)
km.cluster_centers_
```

```
#Elbow Method to find out the value of k 
dist=[]
for i in range(1,11):
  km=KMeans(n_clusters=i, init='random',max_iter=100,tol=0.001,random_state=0)
  km.fit(x)
  dist.append(km.inertia_)
plt.plot(range(1,11),dist,marker='*')
plt.xlabel('Number of cluster')
plt.ylabel('Distortion')
```

Conclusion: 
1) With the increase in standard deviation, the number of iterations required to convulged the KMeans algorithm increases. This is because increase in standard devaition spreads the data points in the cluster, resulting in overlapping of clusters. 
2) With each iteration the location of centroid changes which is based on redistribution of the data points; for the given dataset with standard deviation = 0.8, the locations of centroid doesn't change after third iteration which indicates that the algorithm is convulged.
3)KMeans algorith is tested fo k= 1 to 10, where it is observed that after k=3, there is no appreciable reduction in distortion. Therefore, optimum value of k=3 is selected.
