first four charachters of the file id is the year we will set as the field to search our targets. 

```
table = nltk.ConditionalFreqDist((target, fileid[:4]) #The Conditional Frequency Distribution creates a table for searching the target in our fileid 
for fileid in inaugural.fileids() #we remove the stop words from our text files when we use the CFD function
for w in inaugural.words(fileid) #w will store the words we put in target
for target in ['america', 'citizen','war'] #setting targets to be searched   
if w.lower().startswith(target)) #w stores the target words, converts them into lower case 
table.tabulate(conditions=['america', 'citizen','war'], samples=['1841', '1993','2009']) #sample here indicates the years or the fileids in which our targets will be searched.
```
